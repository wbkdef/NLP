```python
import nltk
from nltk.corpus import treebank
from nltk.grammar import ContextFreeGrammar, Nonterminal


list_rules=[]
from collections import defaultdict
d=defaultdict(int)
with open("atis3.treebank", "r") as f:
    for line in f:
        tree=nltk.Tree.parse(line)
        nltk.treetransforms.chomsky_normal_form(tree)
        rules=tree.productions()
        for r in rules:
            d[r]+=1
#             print 1
#         list_rules.extend(rules)
#         print type(list_rules[0])
#         print list_rules[0]
#         print list_rules[0].lhs()
        
#         break

set_rules=set(d.keys())
# print d
set_lhs=set(r.lhs() for r in set_rules) # The set of non-terminals
set_nonterm_rules=set(r for r in set_rules if r.rhs()[0] in set_lhs)
set_term_rules=set_rules-set_nonterm_rules
print "set_lhs"
print set_lhs
print "set_nonterm_rules"
print set_nonterm_rules
print "set_term_rules"
print set_term_rules


```
